```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv(earthquake_data1)
print('Shape: ', data.shape, '\n')
```

```python
data['date_time'] = pd.to_datetime(data['date_time'])

# Splitting into date and time columns
data['Date'] = data['date_time'].dt.date
data['Time'] = data['date_time'].dt.time
```

```python
# List of columns to remove
columns_to_remove = ['title', 'date_time', 'mmi', 'net', 'nst', 'sig', 'gap', 'magType']

# Drop the specified columns
data = data.drop(columns=columns_to_remove)
```

```python
column_mapping = {
    'magnitude': 'Magnitude',
    'alert': 'Alert',
    'tsunami': 'Tsunami',
    'depth': 'Depth',
    'latitude': 'Latitude',
    'longitude': 'Longitude',
    'continent': 'Continent',
    'location': 'Location',
    'dmin': 'Epicenter',
    'cdi': 'Intensity',
    'country': 'Country',
}

# Rename the columns using the mapping
data = data.rename(columns=column_mapping)
```

```python
column_order = ['Date', 'Time', 'Continent', 'Location', 'Country', 'Latitude', 'Longitude', 'Magnitude', 'Depth', 'Intensity', 'Epicenter', 'Alert', 'Tsunami']

# Reorder the columns
data = data[column_order]
data.head()
```

```python
cleaned_data = data.dropna()
cleaned_data.to_csv(earthquake_data2, index=False)
print('File created!')
```

# **Map**

```python
import folium
from folium import plugins
```

```python
df = pd.read_csv(earthquake_data2)
print('Shape: ', df.shape)
df.head()
```

```python
df.tail()
```

```python
# Create a folium map centered at the mean latitude and longitude
map1 = folium.Map(location=[df['Latitude'].mean(), df['Longitude'].mean()], zoom_start=2)

# Add markers with tooltips and circle markers indicating magnitude
for index, row in data.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=f"{row['Location']} - Magnitude: {row['Magnitude']}",
    ).add_to(map1)

    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=row['Magnitude'],  # You can adjust the multiplier for a better visualization
        color='blue',
        fill=True,
        fill_color='blue',
        fill_opacity=0.6,
        popup=f"{row['Location']} - Magnitude: {row['Magnitude']}",
    ).add_to(map1)

map1.save(mappy)  # Save the map to an HTML file
display(map1)
```

```python
print('Earthquake Filtered Data')
# Set thresholds for magnitude and depth
magnitude_threshold = float(input("\nEnter the magnitude threshold: "))
depth_threshold = float(input("\nEnter the depth threshold: "))

# Filter the DataFrame based on the thresholds
filtered_data = df[(df['Magnitude'] >= magnitude_threshold) & (df['Depth'] >= depth_threshold)]
num_earthquakes = len(filtered_data)
print('\nNumber of Earthquakes: ', num_earthquakes)
filtered_data
```

```python
# Create a folium map centered at the mean latitude and longitude
map2 = folium.Map(location=[df['Latitude'].mean(), df['Longitude'].mean()], zoom_start=2)

# Add markers with tooltips for filtered locations
for index, row in filtered_data.iterrows():
    popup_content = f"{row['Location']} - Magnitude: {row['Magnitude']}, Depth: {row['Depth']}, Date: {row['Date']}, Time: {row['Time']}"
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=popup_content,
        icon=folium.Icon(color='red', icon='info-sign')
    ).add_to(map2)

# Save the map to an HTML file
map2.save(heavyearthquakes)

# Display the map
display(map2)
```
